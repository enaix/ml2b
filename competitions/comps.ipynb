```python
import pandas as pd
from pathlib import Path

def add_competition_id_to_all(directory_path, competition_ids):
    """
    Добавляет столбец competition-id ко всем CSV файлам
    competition_ids - список ID для добавления
    """
    folder = Path(directory_path)
    
    for csv_file in folder.glob("*.csv"):
        try:
            df = pd.read_csv(csv_file, sep=None, engine='python')
            
            # Проверяем, что список ID подходит по длине
            if len(competition_ids) != len(df):
                print(f"⚠️  Размер списка ID ({len(competition_ids)}) не совпадает с "
                      f"количеством строк ({len(df)}) в {csv_file.name}")
                continue
            
            # Добавляем competition-id как первый столбец
            df.insert(0, 'competition-id', competition_ids)
            
            df.to_csv(csv_file, index=False, encoding='utf-8')
            print(f"✅ Добавлен competition-id в {csv_file.name}")
            
        except Exception as e:
            print(f"❌ Ошибка с {csv_file.name}: {e}")

# Использование

add_competition_id_to_all("tasks", comp_ids)
```
